```python
import seaborn as sns
import pandas as pd
url = "https://raw.githubusercontent.com/chiharuishida/Tutorialdata/main/ad_dataset_withheaders.csv"
daily = pd.read_csv(url)
daily.columns = ['Site_Time', 'Age', 'DailyTime', 'Gender', 'AdClicking']
daily.head(10)
```

```python
daily.describe()
```

```python
daily.info()
```

```python
daily['Gender'].value_counts()
```

```python
daily.corr()
```

```python
sns.heatmap(daily.corr(), annot=True)
```

```python
sns.heatmap(daily.corr(), cmap="YlGnBu")
```

```python
sns.pairplot(daily)
```

```python
sns.scatterplot(x="Site_Time", y="Age", data=daily)
```

```python
sns.lmplot(x="Site_Time", y="Age", data=daily)
```

```python
sns.scatterplot(x="Site_Time", y="Age", hue="Gender", data=daily)
```

```python
sns.implot(x="Site_Time", y="Age", hue="Gender", data=daily)
```

```python
sns.countplot(x='Gender', data=daily)
```

```python
sns.countplot(x='Gender', data=daily, color='b')
```
